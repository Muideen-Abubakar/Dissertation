```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Import the sentiment scores

agric_I = pd.read_csv('Agric1.csv')
agric_II = pd.read_csv('Agric2.csv')
#merge
agric_sentiment = pd.concat([agric_I, agric_II], axis = 0, ignore_index = True)

energy_I = pd.read_csv('Energy1.csv')
energy_II = pd.read_csv('Energy2.csv')
#merge
energy_sentiment = pd.concat([energy_I, energy_II], axis = 0, ignore_index = True)

industrials_I = pd.read_csv('Industrials1.csv')
industrials_II = pd.read_csv('Industrials2.csv')
#merge
industrials_sentiment = pd.concat([industrials_I, industrials_II], axis = 0, ignore_index = True)

realestate_I = pd.read_csv('Realestate1.csv')
realestate_II = pd.read_csv('Realestate2.csv')
#merge
realestate_sentiment = pd.concat([realestate_I, realestate_II], axis = 0, ignore_index = True)

transport_I = pd.read_csv('Transport1.csv')
transport_II = pd.read_csv('Transport2.csv')
#merge
transport_sentiment = pd.concat([transport_I, transport_II], axis = 0, ignore_index = True)
```

```python
#Import the sentence files 
agric_sentence_I = pd.read_csv('/content/Sentences_Agric1.csv')
agric_sentence_II = pd.read_csv('/content/Sentences_Agric2.csv')
#merge
agric_sentence = pd.concat([agric_sentence_I, agric_sentence_II], axis = 0, ignore_index = True)

energy_sentence_I = pd.read_csv('/content/Sentences_Energy1.csv')
energy_sentence_II = pd.read_csv('/content/Sentences_Energy2.csv')
#merge
energy_sentence = pd.concat([energy_sentence_I, energy_sentence_II], axis = 0, ignore_index = True)

industrials_sentence_I = pd.read_csv('/content/Sentences_Industrials1.csv')
industrials_sentence_II = pd.read_csv('/content/Sentences_Industrials2.csv')
#merge
industrials_sentence = pd.concat([industrials_sentence_I, industrials_sentence_II], axis = 0, ignore_index = True)

realestate_sentence_I = pd.read_csv('/content/Sentences_Realestate1.csv')
realestate_sentence_II = pd.read_csv('/content/Sentences_Realestate2.csv')
#merge
realestate_sentence = pd.concat([realestate_sentence_I, realestate_sentence_II], axis = 0, ignore_index = True)

transport_sentence_I = pd.read_csv('/content/Sentences_Transport1.csv')
transport_sentence_II = pd.read_csv('/content/Sentences_Transport2.csv')
#merge
transport_sentence = pd.concat([transport_sentence_I, transport_sentence_II], axis = 0, ignore_index = True)
```

```python
#Merge Sentence files and sentiments df
agric_score = pd.merge(agric_sentence, agric_sentiment, left_on = 'Sentence', right_on = 'sentence', how = 'right' )
energy_score = pd.merge(energy_sentence, energy_sentiment, left_on = 'Sentence', right_on = 'sentence', how = 'right' )
industrials_score = pd.merge(industrials_sentence, industrials_sentiment, left_on = 'Sentence', right_on = 'sentence', how = 'right' )
realestate_score = pd.merge(realestate_sentence, realestate_sentiment, left_on = 'Sentence', right_on = 'sentence', how = 'right' )
transport_score = pd.merge(transport_sentence, transport_sentiment, left_on = 'Sentence', right_on = 'sentence', how = 'right' )

#Drop duplicated sentence column
agric_score = agric_score.drop('sentence', axis = 1)
energy_score = energy_score.drop('sentence', axis = 1)
industrials_score = industrials_score.drop('sentence', axis = 1)
realestate_score = realestate_score.drop('sentence', axis = 1)
transport_score = transport_score.drop('sentence', axis = 1)

#write a code to drop duplicate rows

#Drop Nan values
agric_score = agric_score.dropna()
energy_score = energy_score.dropna()
industrials_score = industrials_score.dropna()
realestate_score = realestate_score.dropna()
transport_score = transport_score.dropna()
```

```python
#Import the cleaned csvs after filtering out neutral texts
agric_score_cleaned = pd.read_csv('/content/agric_score_exneutral.csv',encoding= 'unicode_escape')
energy_score_cleaned = pd.read_csv('/content/energy_score_exneutral.csv',encoding= 'unicode_escape')
industrials_score_cleaned = pd.read_csv('/content/industrials_score_exneutral.csv',encoding= 'unicode_escape')
realestate_score_cleaned = pd.read_csv('/content/realestate_score_exneutral.csv',encoding= 'unicode_escape')
transport_score_cleaned = pd.read_csv('/content/transport_score_exneutral.csv',encoding= 'unicode_escape')
```

```python
#Convert the Year column to an integer
agric_score_cleaned['Year'] = agric_score_cleaned['Year'].astype(int)
energy_score_cleaned['Year'] = energy_score_cleaned['Year'].astype(int)
industrials_score_cleaned['Year'] = industrials_score_cleaned['Year'].astype(int)
realestate_score_cleaned['Year'] = realestate_score_cleaned['Year'].astype(int)
transport_score_cleaned['Year'] = transport_score_cleaned['Year'].astype(int)
```

```python
#After exporting the dfs above as CSv and manually cleaning them, import them back and add _clean as suffix to their file names for the rest of codes below
```

```python
#Group by companies and year to get the average sentiment score
agric_score_grouped = agric_score_cleaned.groupby(['Company', 'Year']).mean()
energy_score_grouped = energy_score_cleaned.groupby(['Company', 'Year']).mean()
industrials_score_grouped = industrials_score_cleaned.groupby(['Company', 'Year']).mean()
realestate_score_grouped = realestate_score_cleaned.groupby(['Company', 'Year']).mean()
transport_score_grouped = transport_score_cleaned.groupby(['Company', 'Year']).mean()
```

```python
transport_score_grouped = transport_score_grouped.iloc[:30] #drop UPS from the list, (it contained only four years)
energy_score_grouped = energy_score_grouped.drop(index = 'NextEra Energy Inc', axis = 0) #drop Nextera since all the text for 2017 were neutral
```

```python
#Import the cost of capital
url = 'https://github.com/Muideen-Abubakar/Dissertation/blob/main/Cost_of_Capital.xlsx?raw=true'
agric_wacc = pd.read_excel(url, sheet_name = 'Agric_and_Cons_Staple')
energy_wacc = pd.read_excel(url, sheet_name='Energy')
industrials_wacc = pd.read_excel(url, sheet_name = 'Industrial')
realestate_wacc = pd.read_excel(url, sheet_name = 'R_Estate')
transport_wacc = pd.read_excel(url, sheet_name = 'Transportation')
```

```python
#Drop rows of company that are missing in the grouped df
agric_wacc = agric_wacc.drop(7, axis =0) #drop Kellog's
energy_wacc = energy_wacc.drop(1, axis = 0) # drop Nextera since it was deleted due to having all neutral sentence in 2017
industrials_wacc = industrials_wacc.drop(0, axis= 0) #drop Linde Plc
transport_wacc = transport_wacc.drop([0,2], axis= 0) #drop tsla and UPS
```

```python
#sort the Ticker column in alphabetical order
agric_wacc = agric_wacc.sort_values('Company', ascending = True, ignore_index = True)
energy_wacc = energy_wacc.sort_values('Company', ascending = True, ignore_index = True)
industrials_wacc = industrials_wacc.sort_values('Company', ascending = True, ignore_index = True)
realestate_wacc = realestate_wacc.sort_values('Company', ascending = True, ignore_index = True)
transport_wacc = transport_wacc.sort_values('Company', ascending = True, ignore_index = True)
```

```python
industrials_wacc
```

```python
#Convert the wacc to lists
agric_wacc_list = agric_wacc.loc[:,2017:2021].values.flatten()
energy_wacc_list = energy_wacc.loc[:,2017:2021].values.flatten()
industrials_wacc_list = industrials_wacc.loc[:,2017:2021].values.flatten()
realestate_wacc_list = realestate_wacc.loc[:,2017:2021].values.flatten()
transport_wacc_list = transport_wacc.loc[:,2017:2021].values.flatten()

#df.to_numpy().flatten()
```

```python
agric_score_grouped['WACC'] = agric_wacc_list
energy_score_grouped['WACC'] = energy_wacc_list
industrials_score_grouped['WACC'] = industrials_wacc_list
realestate_score_grouped['WACC'] = realestate_wacc_list
transport_score_grouped['WACC'] = transport_wacc_list
```

```python
transport_score_grouped
```

```python
#Display the correlation coefficients in tabular form per company

agric_correlation = round(np.corrcoef(agric_score_grouped['sentiment_score'], agric_score_grouped['WACC'])[0,1],3)

energy_correlation = round(np.corrcoef(energy_score_grouped['sentiment_score'], energy_score_grouped['WACC'])[0,1],3)

industrials_correlation = round(np.corrcoef(industrials_score_grouped['sentiment_score'], industrials_score_grouped['WACC'])[0,1],3)

realestate_correlation = round(np.corrcoef(realestate_score_grouped['sentiment_score'], realestate_score_grouped['WACC'])[0,1],3)

transport_correlation = round(np.corrcoef(transport_score_grouped['sentiment_score'], transport_score_grouped['WACC'])[0,1],3)
```

```python
#Plot the correlation 

fig, axs = plt.subplots(3,2, figsize = [20,15], constrained_layout=True)

axs[0,0].scatter(agric_score_grouped['sentiment_score'], agric_score_grouped['WACC'], label = 'Corr = '+str(agric_correlation) )
axs[0,0].set_title('Agric Correlation')                                               
axs[0,0].set_xlabel("Sentiment Score")
axs[0,0].set_ylabel("WACC")
axs[0,0].legend()

axs[0,1].scatter(energy_score_grouped['sentiment_score'], energy_score_grouped['WACC'], label = 'Corr = '+str(energy_correlation))
axs[0,1].set_title('Energy/ELectric Power Correlation')
axs[0,1].set_xlabel("Sentiment Score")
axs[0,1].set_ylabel("WACC")
axs[0,1].legend()

axs[1,0].scatter(industrials_score_grouped['sentiment_score'], industrials_score_grouped['WACC'], label = 'Corr = '+str(industrials_correlation))
axs[1,0].set_title('Industrials Correlation')
axs[1,0].set_xlabel("Sentiment Score")
axs[1,0].set_ylabel("WACC")
axs[1,0].legend()

axs[1,1].scatter(realestate_score_grouped['sentiment_score'], realestate_score_grouped['WACC'], label = 'Corr = '+str(realestate_correlation))
axs[1,1].set_title('Real Estate Correlation')
axs[1,1].set_xlabel("Sentiment Score")
axs[1,1].set_ylabel("WACC")
axs[1,1].legend()

axs[2,0].scatter(transport_score_grouped['sentiment_score'], transport_score_grouped['WACC'], label = 'Corr = '+str(transport_correlation))
axs[2,0].set_title('Transport Correlation')
axs[2,0].set_xlabel("Sentiment Score")
axs[2,0].set_ylabel("WACC")
axs[2,0].legend()

fig.delaxes(axs[2,1])
```
